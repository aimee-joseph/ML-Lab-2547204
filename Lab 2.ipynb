# Task 2

```python
# AQI and AQI_Bucket are the mostly the target variables for this project. It doesn't make sense to to have rows without their target value. 

city_day_clean = city_day.dropna(subset = ['AQI', 'AQI_Bucket']).copy()
```

```python
# Filling the pollutant columns with the median value for each city.

pollutant_cols = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']

city_day_clean[pollutant_cols] = (city_day_clean.groupby('City')[pollutant_cols].transform(lambda x: x.fillna(x.median())))
```

```python
# There might still be missing values, so filling them with the overall median for each pollutant.

city_day_clean[pollutant_cols] = city_day_clean[pollutant_cols].fillna(city_day_clean[pollutant_cols].median())
```

```python
city_day_clean.isnull().sum()
```

For the pollutant columns, I filled the missing values with the city-wise median, as the dataset in general, has a lot of outliers. Using the mean, implies that every value is used directly, so a few large values can distort it.

The median represents the typical value because it is the middle value after sorting.

```python
crop_production_clean = crop_production.copy()

crop_production_clean['Production'] = (crop_production_clean.groupby(['State_Name', 'Crop'])['Production'].transform(lambda x: x.fillna(x.median())))
```

```python
crop_production_clean['Production'] = (crop_production_clean.groupby('Crop')['Production'].transform(lambda x: x.fillna(x.median())))
```

```python
crop_production_clean.isnull().sum()
```

I filled the Production column with median values again, because the values seem to be extreme.

```python
city_day_clean_treated['Date'].info()
```

```python
# extracting the year for year-wise analysis

city_day_clean_treated['Year'] = city_day_clean_treated['Date'].dt.year
city_day_clean_treated['Year'].head()
```

```python
yearly_aqi = city_day_clean_treated.groupby('Year')['AQI'].mean().reset_index()
print("Mean AQI for each year")
yearly_aqi
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly_aqi, x='Year', y='AQI', marker='o')
plt.title('Average AQI by Year')
plt.xlabel('Year')
plt.ylabel('Mean AQI')
plt.tight_layout()
plt.show()
```

We can see a decline in the average AQI over the 5 years, implying that the air quality seems to have improved, with it being at above 200 in 2015, to below 120 in 2020.

```python
# since city_day_clean_treated has city-wise daily records, while crop_production_clean has state-wise yearly records, we cannot merge them directly

# since we mapped cities to states earlier in Task 3, we can get the state-wise yearly average AQI and then merging it with the crop production data

aqi_by_state_year = city_day_clean_treated.groupby(['State_Name', 'Year'])['AQI'].mean().reset_index()
aqi_by_state_year.head()
```

```python
aqi_by_state_year['State_Name'].unique()
```

```python
aqi_by_state_year['Year'].unique()
```

```python
crop_by_state_year = crop_production_clean.groupby(['State_Name', 'Crop_Year'])['Production'].sum().reset_index().rename(columns={'Crop_Year': 'Year'})
crop_by_state_year.head()
```

```python
crop_by_state_year['State_Name'].unique()
```

```python
crop_by_state_year['Year'].unique()
```

```python
merged = aqi_by_state_year.merge(crop_by_state_year, on=['State_Name', 'Year'], how='inner')
merged.head()
```

There are no common state + year combination values in both datasets, leading to an empty dataframe being displayed.

Since the task specifically requests for comparison of states with worse air quality and their crop production, I'll be grouping and merging them based on state name alone.

```python
aqi_by_state = city_day_clean_treated.groupby('State_Name')['AQI'].mean().reset_index()
aqi_by_state.head()
```

```python
crop_by_state = crop_production_clean.groupby('State_Name')['Production'].sum().reset_index()
crop_by_state.head()
```

```python
merged_state = aqi_by_state.merge(crop_by_state, on='State_Name', how='inner')
merged_state.head()
```

```python
corr = merged_state[['AQI', 'Production']].corr()
corr
```

```python
corr = merged_state.select_dtypes(include='number').corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Combined Dataset')
plt.tight_layout()
plt.show()
```

As per the data in the merged dataframe, there is a negative correlation of -0.21 between AQI and Production, which implies that as AQI goes up, Production tends to go down.
